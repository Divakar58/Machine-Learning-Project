```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
dtypes={
    'ip':'uint16',
    'app':'uint16',
    'device':'uint16',
    'os':'uint16',
    'channel':'uint16',
    'click_id':'uint32',
    'is_attributed':'uint8'
}
testing=True
if testing:
    train_path='train_sample.csv'
    skiprows=None
    nrows=None
    columns=['ip','app','device','os', 'channel', 'click_time', 'is_attributed']
else:
    train_path='train.csv'
    skiprows=range(1,144903891)
    nrows=10000000
    columns=['ip','app','device','os', 'channel', 'click_time', 'is_attributed']
click_df=pd.read_csv(train_path,skiprows=skiprows,nrows=nrows,dtype=dtypes,usecols=columns)
```
